```python
# First, we will make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices= "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit.

```python
# Our dataframe lists 48 boroughs, many of which are indeed not London Boroughs, like 'East of England' and 'London'
# We will drop these, but first lets look at the Unnamed Boroughs and see if they contain any useful information

print(clean_properties[clean_properties['London_Borough'] == 'Unnamed: 34'].head())
print(clean_properties[clean_properties['London_Borough'] == 'Unnamed: 37'].head())
print(clean_properties[clean_properties['London_Borough'] == 'Unnamed: 47'].head())
```

```python
# All of this Unnmamed have null values, lets drop all the rows that have null values
# There is no way to immute the values from the information contained here.

NaNfreeDF = clean_properties.dropna()
NaNfreeDF
```

```python
#Now let's drop the remaining non London Boroughs, here is a list of there names that we will use to drop from.

nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']

# We will filter out these boroughs by only keeping boroughs NOT in this list

NaNfreeDF = NaNfreeDF[~NaNfreeDF.London_Borough.isin(nonBoroughs)]

print(NaNfreeDF.London_Borough.unique(), '\n')
print("Number of boroughs remaining in dataframe: ", NaNfreeDF.London_Borough.nunique())
```

```python
# Great, now we have 32 boroughs plus one extra for the whole City of 'London'.
# Next I will analyze the the data with some visualizations.
# Before that I will rename and copy the dataframe indicating that Data Cleaning is complete
# I will also add a month and year column to make data analysis a little cleaner to index and slice.

df = NaNfreeDF.copy()
df['Month'] = df['Date'].apply(lambda d: d.month)
df['Year']= df['Date'].apply(lambda d: d.year)
df.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
# I will first find the average price per year for each borough.
# I will do this by using a groupby function, grouping by borough and year, and then resetting the index.

df_grouped = df.groupby(by=['London_Borough', 'Year']).mean().drop('Month', axis=1)
df_grouped = df_grouped.reset_index()
df_grouped.sample(10)
```

```python
# Next I will write a function that will calculate the change in price from 1998 until 2018.

def price_change(bor, first_year = 1998, last_year = 2018):
    first_yr_avg = float(bor['Average_price'][bor['Year']==first_year])
    last_yr_avg = float(bor['Average_price'][bor['Year']== last_year])
    return last_yr_avg / first_yr_avg
```

```python
# First I will instantiate an empty dictionary to store all the price change ratios of each neighborhood.

borough_price_change_dict = {}

# Now I will loop through all the boroughs and call the price_change function on each iteration, storing the result
# in the dictionary.

for b in list_of_boroughs:
    borough = df_grouped[df_grouped['London_Borough'] == b]
    ratio = price_change(borough )
    borough_price_change_dict[b] = [ratio]
    
print(borough_price_change_dict)
```

```python
# Lets convert this dictionary into a dataframe for easier analysis.

ratio_df = pd.DataFrame(borough_price_change_dict)
ratio_df
```

```python
# I will transpose and reset the index so we can analyze the borough alongside the change in price.
# I will also name the column headers

ratio_df = ratio_df.T.reset_index().rename(columns={'index':'Borough', 0:'Change in Price'})
ratio_df.head()
```

```python
# I will graph all boroughs to compare 
ratio_df =ratio_df.sort_values('Change in Price',ascending=False)
ax = ratio_df[['Borough','Change in Price']].plot(kind='bar')

ax.set_xticklabels(ratio_df.Borough)
plt.show()
```

```python
top5 = ratio_df.head()
top5

# Here we can see the top 5 boroughs that have increased in price the most in this twenty year span.
# They increased approximately between 5.3-6.2 times in values.
```

```python
bottom5 = ratio_df.tail()
bottom5

# Here we see the five boroughs with the slowest growth.
# They still approximetaley quadrupled in value in this period.
```
