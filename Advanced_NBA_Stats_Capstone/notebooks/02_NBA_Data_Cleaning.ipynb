# 2. Data Cleaning

- In this notebook we will perform data cleaning. The column names need to be reformatted. A small number of columns need to converted into different file types and there are a small number of redundant columns that were scraped from multiple webpages. The data frame with need some additional wrangling after EDA, once colinear columns are analyzed.

```python
# importing packages
import pandas as pd 
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('data/scaped_nba_stats_data.csv')
```

__Cleaning Column Names__

```python
# There are some whitespaces between some of the column names. Let's remove those first.

data.columns = data.columns.str.replace(' ', '_').str.replace('\xa0', '_')
data.head()
```

__Investigating Duplicate Columns/NaN values in columns.__

```python
# fishy looking columns ['AST_PTSCreated	AST_PTSCreated.1	AST_PTSCreated.2	AST_PTSCreated.3']
# .1 and .3 contain the same information, whereas .0 and .2 are only null values.

data[['AST_PTSCreated', 'AST_PTSCreated.1', 'AST_PTSCreated.2', 'AST_PTSCreated.3']]
```

```python
# Let's keep only one of these columns and rename it.

data.drop(['AST_PTSCreated', 'AST_PTSCreated.2', 'AST_PTSCreated.3'],axis=1, inplace=True)
data.rename(columns={'AST_PTSCreated.1':'AST_PTS_Created'},inplace=True)
```

```python
data.head()
```

```python
#Here we can see the the columns with an _x and _y suffix are duplicates. I will drop one and rename the other.

# Draft Round is redundant as the first 30 draft numbers are from round 1, the last 30 are from round 2.
# I will drop Draft Round and Draft Year as well because it is redundant to the players age (not a great way to impute null values without just adding an Experience column)(might later add a column of years in the league)

data[['AST_x','AST_y','Draft_Year_x','Draft_Round_x', 'Draft_Number_x','Height_y','Weight_y','College_y','Country_y','Draft_Year_y','Draft_Round_y','Draft_Number_y']]
```

```python
data.drop(['AST_y','Draft_Year_x','Draft_Round_x','Height_y','Weight_y','College_y','Country_y','Draft_Year_y','Draft_Round_y','Draft_Number_y'],axis=1,inplace=True)
data.rename(columns={'AST_x':'AST','Height_x':'Height', 'Weight_x':'Weight', 'College_x':'College', 'Country_x':'Country','Draft_Number_x':'Draft_Number'},inplace=True)
```

```python
data.head()
```

__Converting Height column to a single integer in inches instead of being in feet and inches.__

```python
#function to map to Height Series to convert Feet/Inches into Total Inches.

def ft_in_to_inches(ft_in_str):
    ht_tuple = ft_in_str.split('-')
    ft_ = float(ht_tuple[0])
    in_ = float(ht_tuple[1])
    return (ft_ * 12) + in_
    
data.Height = data.Height.apply(lambda x: ft_in_to_inches(x))
data.Height
```

__Dropping Rows for Players that did not play in the 2018 Season.__

This is our target feature that we are trying to predict. Also the reason why these values are null is because these are players taht did not play in the 2018 (retired). The model we are tying to train is to predict forecast values
across seasons, so player continuity is necessary.

```python
print('# of players that played in the 2017 but not in 2018:',data.MIN_2018.isnull().sum())
data[data.MIN_2018.isnull()].head(3)
```

```python
data = data[data.MIN_2018.notnull()]
data.shape
```

__The Draft_Number column has non-numeric values('Undrafted'). For modelling purposes we need this to be numeric.
In the NBA draft there are 60 picks. For simplicity all undrafted players we will recode them with a draft position 
of 61 (https://en.wikipedia.org/wiki/2017_NBA_draft).__

```python
data.Draft_Number.value_counts()[:5]
```

```python
data.Draft_Number = data.Draft_Number.str.replace('Undrafted', '61')
data.Draft_Number = data.Draft_Number.astype(int)
data.Draft_Number.value_counts()[:5]
```

```python
data.dtypes.value_counts()
```

__Changing the Datatype for the rest of the Dataframe to numeric (excluding actual strings)__

```python
cols_str_vals = data.select_dtypes('object').drop(['Player','TEAM', 'College','Country'],axis=1).columns

mask = np.column_stack([data[col].str.contains(r"-", na=False) for col in data[cols_str_vals] ])
data[cols_str_vals].loc[mask.any(axis=1)]

# there are a number of values in the dataframe that are listed as '-'. 
# These are null values. We will change them to '0.0'.
```

```python
#data[cols_str_vals] = [col.str.replace('-','0.0') for col in data[cols_str_vals]]
data[cols_str_vals] = data[cols_str_vals].replace(['-'],'0.0')
```

```python
mask = np.column_stack([data[col].str.contains(r"-", na=False) for col in data[cols_str_vals] ])
print("Number of occurences of '-':", data[cols_str_vals].loc[mask.any(axis=1)].sum().sum())
```

```python
# Now changing columns with object values to floats

data[cols_str_vals] = data[cols_str_vals].astype('float')
```

```python
# Below we can see that there are only four columns with object values, and as you can see in the filtered dataframe
# below these columns indeed should be this dtype.

print(data.dtypes.value_counts())
data.select_dtypes('object')
```
